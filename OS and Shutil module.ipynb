# Shutil module in python

The Shutil module in Python provides many functions of high-level operations on files and collections of files. In particular, functions are provided which support file `copying` and `removal`.

## Move

```python
import shutil
# 1- create new directory (folder) to move the file 'test.txt' to it
# 2- using shutil.move move the file, then list the content of the new folder
os.mkdir('New Folder')
```

```python
shutil.move('test.txt', '/content/New Folder')
```

```python
os.listdir('New Folder')
```

```python
os.listdir()  # 'test.txt' is removed
```

## Copy

```python
# You can also copy the test.txt file using the shutil module:
shutil.copy('/content/New Folder/test.txt', '/content')
```

## Delete

The os module provides 3 methods to delete the files:

os.uplink(‘path’) which deletes a file at the path you provided

os.rmdir(‘path’) which deletes the folder (folder should be empty for this to work) at the path you provided.

shutil.rmtree(‘path’) which deletes all the files and folders contained in the path you provided.

```python
shutil.rmtree('/content/New Folder')
```

## Shutil.copyfile
shutil.copyfile() method in Python is used to copy the content of the source file to the destination file. The metadata of the file is not copied. Source and destination must represent a file and destination must be writable. If the destination already exists then it will be replaced with the source file otherwise a new file will be created.

```python
shutil.copy('test.txt', '/content/sample_data')
```
